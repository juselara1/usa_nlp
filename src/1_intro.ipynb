```python
from typing import Dict, List, Type
from pathlib import Path
from dataclasses import dataclass

DataType = int | float | str

@dataclass
class Row:
    fields: Dict[str, DataType]

@dataclass
class Table:
    rows: List[Row]
```

```python
Schema = Dict[str, Type]
def parse_row(line: str, schema: Schema) -> Row | None:
    try:
        vals = line.split('"')
        all_vals = []
        all_vals.extend(vals[0][:-1].split(","))
        all_vals.append(vals[1])
        all_vals.extend(vals[2][1:].split(","))
        types = schema.values()
        parsed_vals = list(map(lambda cats: cats[0](cats[1]), zip(types, all_vals)))
        return Row(
            fields = dict(zip(schema.keys(), parsed_vals))
        )
    except:
        return None

parse_row('1,0,3,"Braund, Mr. Owen Harris",male,22,1,0,A/5 21171,7.25,,S', schema)
```

```python
def load_csv(path: Path, schema: Schema) -> Table:
    with open(path, 'r') as f:
        data = f.read()
    lines = data.split("\n")
    rows = map(lambda line: parse_row(line, schema), lines)
    clean_rows = list(filter(lambda row: row is not None, rows))
    return Table(rows=clean_rows)
```

```python
path = Path("data.csv")
schema = {
    "PassengerId": int,
    "Survived": bool,
    "Pclass": int,
    "Name": str,
    "Sex": str,
    "Age": int,
    "SibSp": int,
    "Parch": int,
    "Ticket": str,
    "Fare": float,
    "Cabin": str,
    "Embarked": str
}
load_csv(path, schema)
```
