```python
isletter_fn = lambda x: x >= 'A' and x <= 'Z'
isnum_fn = lambda x: x >= '0' and x <= '9'
def validate(text: str) -> bool:
    if len(text) != 6: return False
    letters = text[:3]
    all_letters = all(map(isletter_fn, letters))

    nums = text[3:]
    all_nums = all(map(isnum_fn, nums))

    return all_letters and all_nums
```

aetuhateohu@correo.com -> aetuhasteuh@correo.es

tuhaeotuhae@correo.es -> tauheotnashu@correo.lat


```python
def modify_domain(mat: re.Match) -> str:
    user = mat.group("user")
    domain = mat.group("domain")
    if domain == "correo.es":
        new_domain = "correo.lat"
    elif domain == "correo.com":
        new_domain = "correo.es"
    else:
        new_domain = domain
    return "@".join((user, new_domain))
```

```python
x = "el correo de pepe@correo.es junto con el correo de maria@correo.com pero no el correo marta@correo.io"
```

```python
pat = re.compile(r"(?P<user>\w+)@(?P<domain>[\w\.]+)")
```

```python
re.sub(pat, modify_domain, x)
```
