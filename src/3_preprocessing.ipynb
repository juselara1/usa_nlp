```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("https://es.wikipedia.org/wiki/Cross_Industry_Standard_Process_for_Data_Mining")
```

```python
assert r.status_code == 200
```

```python
text = r.text
```

```python
raw_data = (
    BeautifulSoup(text)
    .find(id="bodyContent")
    .get_text()
)
```

```python
print(raw_data)
```

- **Token**: minima unidad de procesamiento del modelo (caracter, palabra, oracion, prefijo y sufijo, ...)
- **Vocabulario**: tokens unicos en el corpus.

### NLTK

```python
from nltk.corpus import stopwords
```

```python
sw = stopwords.words("spanish")
```

```python
word_tokens = map(lambda tok: tok.lower(), word_tokenize(raw_data))
```

```python
filtered_tokens = list(filter(lambda tok: tok not in sw, word_tokens))
```

```python
counts = (
    pd.Series(filtered_tokens)
    .value_counts()
    .reset_index()
    .rename(columns={"index": "word", 0: "count"})
)
```

```python
fig = px.bar(counts, x="word", y="count")
fig.show()
```

### Spacy

```python
doc = nlp(raw_data)
filtered_tokens = filter(lambda tok: not tok.is_stop, doc)
filtered_tokens = list(map(lambda tok: tok.text, filtered_tokens))
print(filtered_tokens)
```

```python
counts = (
    pd.Series(filtered_tokens)
    .value_counts()
    .reset_index()
    .rename(columns={"index": "word", 0: "count"})
)
```

```python
fig = px.bar(counts, x="word", y="count")
fig.show()
```

### Spacy

```python
lemmas = map(lambda tok: tok.lemma_, doc)
lemma_text = " ".join(lemmas)
```

```python
print(lemma_text)
```

```python
filtered_tokens = filter(lambda tok: not tok.is_stop, doc)
lemmas = map(lambda tok: tok.lemma_, filtered_tokens)
lower_lemmas = map(lambda lem: lem.lower(), lemmas)
```

```python
counts = (
    pd.Series(lower_lemmas)
    .value_counts()
    .reset_index()
    .rename(columns={"index": "word", 0: "count"})
)
```

```python
fig = px.bar(counts, x="word", y="count")
fig.show()
```
